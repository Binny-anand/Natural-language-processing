```python
#data preprocessing
```

In this process,we will first replace symbols(?,\,@ etc) with space.
then we will be eliminating non_significantterms like a,an,the with the help stopwords and then stemming

```python
#importing necessary libraries
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
train_corpus=[]
test_corpus=[]
```

```python
for i in range(0,7613):
    text=re.sub('[^a-z A-Z]',' ',train["text"][i])
    text=text.lower()
    text=text.split()
    text=[ps.stem(word) for word in text if word not in set(stopwords.words('english'))]
    text=" ".join(text)
    train_corpus.append(text)
    
```

```python
train_corpus
```

```python
for i in range(0,3263):
    text=re.sub('[^a-z A-Z]',' ',test["text"][i])
    text=text.lower()
    text=text.split()
    text=[ps.stem(word) for word in text if word not in set(stopwords.words('english'))]
    text=" ".join(text)
    test_corpus.append(text)
```

```python
test_corpus
```

```python
#Creating a model
```

we will be creating a sparse matrix in which rows are the texts from our dataset and columns are each word appeared in document

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=3000)
features=cv.fit_transform(train_corpus).toarray()
labels=train.iloc[:,2].values
```

```python
features.shape
```

```python
labels.shape
```

```python
text_to_predict=cv.fit_transform(test_corpus).toarray()
```

```python
#building classifier

from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=8)
classifier.fit(features,labels)
```

```python
test["target"]=classifier.predict(text_to_predict)
```
